# Reading the dataset

```
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
# Reading the dataset
dataset = pd.read_csv('Experience_Salaries.csv')
```

```
'''
Alternatively, this dataset can be read from AIM Mock datasets using the below line of code

data = pd.read_csv(
    'https://raw.githubusercontent.com/analyticsindiamagazine/MocksDatasets/main/Experience_Salaries.csv'
    )
'''
```

```
# Cheking the dataset
print(dataset)
```

# Defining the input-output feature

```
# defining training features (X and y)
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```
# Checking the input feature
print(X)
```

```
# Checking the output feature
print(y)
```

```
# Reshaping the output feature
y = y.reshape(len(y),1)
```

```
# Reshaped feature
print(y)
```

# Feature scaling

```
# Standardiing the input and output features
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```
# Standardized input values
print(X)
```

```
# Standardized output values
print(y)
```

# Defining and training the support vector regression model

```
# Defining the support vector regressor
from sklearn.svm import SVR

# Training the support vector regressor
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y)
```

# Visualising the SVM performance in regression

```
# Performance of regressor on the training data
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'blue')
plt.plot(sc_X.inverse_transform(X), sc_y.inverse_transform(regressor.predict(X)), color = 'red')
plt.title('Experience Vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

# Making predictions with a random input

```
# Predicting salary for a random experience
sc_y.inverse_transform(regressor.predict(sc_X.transform([[3.5]])))
```
