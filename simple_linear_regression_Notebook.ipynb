# Simple Linear Regression

## Importing the libraries

```python
# Importing the required libraies
import numpy as np
import pandas as pd
```

## Importing the dataset

```python
# Reading the dataset
dataset = pd.read_csv('Salary_Data.csv')
```

```python
# Top 5 rows of the data
dataset.head()
```

```python
# Checking the shape of the data
print('Shape of the dataset (No. of rows, No. of columns):', dataset.shape)
```

```python
# Defining input and output features
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Checking the shape of input and output features
print('Shape of the input features:', X.shape)
print('Shape of the output features:', y.shape)
```

## Splitting the dataset into the Training set and Test set

```python
# Defining the training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# Checking the shape of the training and test sets
print('Shape of the training data:', X_train.shape)
print('Shape of the test data:', y_train.shape)
```

## Training the Simple Linear Regression model on the Training set

```python
# Importing the library
from sklearn.linear_model import LinearRegression

#Defining the simple linear regression model
regressor = LinearRegression()

# Training the simple linear regression model
regressor.fit(X_train, y_train)
```

```python
print(regressor)
```

```python
regressor
```

```python
# The regression coefficient
print('The regression coefficient:', regressor.coef_[0])
```

```python
# The intercept
print('The intercept', regressor.intercept_)
```

```python
# The fitted linear regression line
print('The regression line is: y = ', regressor.intercept_, '+', regressor.coef_[0],'*X')
```

```python
pred_0 = regressor.predict([[10]])
print('Predicted value of Y when X = 10 : ', pred_0[0])
```

## Predicting the Test set results

```python
# Making predictions on the test data
y_pred = regressor.predict(X_test)
```

```python
# Comparing the predicted salaries with actual salaries
pd.DataFrame(data={'Predicted Salary': y_pred, 'Actual Salary': y_test})
```

# Visualising the Test set results

```python
# Visualizing the fitted regression line against the predicted values
import matplotlib.pyplot as plt
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test, y_pred)
print('Mean Squared Error is:', MSE)
```

```python
# Root Mean Squared Error (RMSE)
import math
RMSE = math.sqrt(MSE)
print('Root Mean Squared Error is:', RMSE)
```

```python
# R-Squared
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('R-Squared is:', r2)
```
