#Setting Dates

```
import datetime as dt
from datetime import datetime as dt
from dateutil.relativedelta import relativedelta
```

```
print(start)
print(end)
```

#SBI Data

```
data = yf.download('SBIN.NS', start, end)
```

```
data.shape
```

# Day Closing with Linear Regression

```
import numpy as np
```

```
x = data.iloc[:, :1].values
y = data.iloc[:, 3].values
```

```
x
```

```
y
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
```

```
linreg.fit(X_train,y_train)
```

```
y_pred = linreg.predict(X_test)
```

```
y_pred
```

```
column_values = ['Actual', 'Predicted']
```

```
values = np.array([y_test, y_pred])
```

```
import pandas as pd
```

```
res = pd.DataFrame(data={'Predicted': y_pred, 'Actual': y_test})
```

```
res
```

```
res.corr()
```

```
# Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test, y_pred)
print('Mean Squared Error is:', MSE)
```

```
# Root Mean Squared Error (RMSE)
import math
RMSE = math.sqrt(MSE)
print('Root Mean Squared Error is:', RMSE)
```

```
val = 23.75
pred = linreg.predict([[val]])
print(pred)

diff = pred-val
if diff>0:
  print('Gain today')
else:
  print('Loss today')
```
